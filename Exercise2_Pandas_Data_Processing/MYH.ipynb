## MYH data

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("data/resultat-ansokningsomgang-2024.xlsx",sheet_name="Tabell 3", skiprows=5)
df.head(5)
```

```python
utbildningsnamn = df["Utbildningsnamn"].nunique()
utbildningsnamn
```

```python
utbildningsomrade = df["Utbildningsområde"].nunique()
utbildningsomrade
```

```python
data_it = df[df["Utbildningsområde"] == "Data/IT"]
data_it
```

```python
stockholm_providers = data_it[
    data_it["Utbildningsanordnare administrativ enhet"].str.contains(
        "Stockholm", case=False, na=False
    )
]


stockholm_count = stockholm_providers[
    "Utbildningsanordnare administrativ enhet"
].nunique()
stockholm_count
```

```python
gothenburg_providers = data_it[
    data_it["Utbildningsanordnare administrativ enhet"].str.contains(
        "Göteborg", case=False, na=False
    )
]


gothenburg_count = gothenburg_providers[
    "Utbildningsanordnare administrativ enhet"
].nunique()
gothenburg_count
```

```python
total_applied_utbildningsomrade = df.groupby("Utbildningsområde")["Sökta platser totalt"].sum()
print(total_applied_utbildningsomrade)
```

```python
grouped_approved = df.groupby("Utbildningsområde")[["Beviljade platser totalt", "Sökta platser totalt"]].sum()
grouped_approved
```

```python
grouped_approved["success_rate"] = grouped_approved["Beviljade platser totalt"] / grouped_approved["Sökta platser totalt"]
highest_approval = grouped_approved.sort_values("success_rate", ascending= False)
highest_approval.head(5)
```

```python
program_per_field = df.groupby("Utbildningsområde")["Utbildningsnamn"].nunique().reset_index()
plt.figure(figsize=(12,6))
plt.bar(program_per_field["Utbildningsområde"], program_per_field["Utbildningsnamn"])
plt.xticks(rotation=45)
plt.xlabel("Utbildningsområde")
plt.ylabel("Unika program")
plt.title("Utbildningsprogram per område")
plt.tight_layout()

plt.show()
```
